Glob is used to read multiple images from a folder

```python
import cv2
import glob
from matplotlib import pyplot as plt
```

```python
file_list = glob.glob('/content/drive/MyDrive/Images/*')
print(file_list)

#glob.glob returns list of files with their full path
```

**Load Each file at a time**

```python
my_list = []

path = '/content/drive/MyDrive/Images/*'

for file in glob.glob(path):
  print(file)
  a = cv2.imread(file)
  my_list.append(a)

print(my_list)
#here my_list is storing numpy array be reading image that is it store pixels
```

**View Stored Images**

```python
plt.imshow(my_list[2])
```

**Using glob perform some operations on images**

```python
for file in glob.glob(path):
    a= cv2.imread(file)

    c = cv2.cvtColor(a, cv2.COLOR_BGR2RGB)
    img_number = 1
    cv2.imwrite("images/test_images/Color_image"+str(img_number)+".jpg", c)
    img_number += 1
    plt.imshow(c)
```
