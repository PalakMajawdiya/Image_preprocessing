```python
from matplotlib import pyplot as plt

x = [1,2,3,4,5]
y = [1,4,9,16,25]

plt.plot(x,y)
```

plot takes any number of arguments

```python
import numpy as np

a = np.array([1,2,3,4,5])
b = np.array([1,4,9,16,25])
plt.plot(a,b)
plt.plot(a, b, 'bo') #Blue dots. Also try: 'r--' 'g^' 'bs'
plt.axis([0, 6, 0, 50]) #Define range for x and y axes [xmin, xmax, ymin, ymax] 
```

```python
wells = ['well1', 'well2', 'well3', 'well4', 'well5']
cells = [80, 62, 88, 110, 90]

plt.bar(wells, cells)
```

```python
plt.scatter(wells, cells)
```

```python
plt.plot(wells, cells)
```

If you run all lines at once you'll see that all plots share same axes.

```python
plt.bar(wells, cells)
plt.scatter(wells, cells)
plt.plot(wells, cells)
```

```python
a = np.array([1,2,3,4,5])
b = np.array([1,4,9,16,25])
plt.plot(a, b, linewidth=5.0) #for the size of linewidth
```

use setp() to define multiple parameters

```python
fig = plt.plot(a, b, linewidth=5.0)
plt.setp(fig, color='r', linewidth=4.0)
```

Adding labels and annotations

```python
plt.figure(figsize=(8, 8))
plt.bar(wells, cells)
plt.xlabel('Well #', fontsize=18, color='red')
plt.ylabel('# dead cells')
plt.title('Dead cells in each well')
plt.axis([1, 6, 60, 120])  
plt.grid(True)
plt.show()
```

plotting multiple plots side by side
To create plots side by side we use subplot()

```python
x = [1,2,3,4,5]
y = [10, 125, 1350, 11250, 100500]

plt.figure(figsize=(12, 6))

#1st Linear Plot
plt.subplot(121) # (121)Here 1st row 2nd column 1st place 
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)

#2nd Logarithmic Plot
plt.subplot(122) # (122)1st row 2nd column 2nd place 
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)

```

Now Plotting Horizontally


```python
plt.figure(figsize=(16, 6))

#Let us plot in 1 row and 3 columns (13x) x is for the position

plt.subplot(131) 
plt.bar(wells, cells)
plt.subplot(132)
plt.scatter(wells, cells)
plt.subplot(133)
plt.plot(wells, cells)
plt.suptitle('Multiple Plots')
plt.show()
```

Plotting Vertically

```python
plt.figure(figsize=(6, 16))

# Let us plot in 1 column and 3 rows (31x)
plt.subplot(311) 
plt.bar(wells, cells)
plt.subplot(312)
plt.scatter(wells, cells)
plt.subplot(313)
plt.plot(wells, cells)
plt.suptitle('Multiple Plots')
plt.show()
```

Plotting as a Grid

```python
plt.figure(figsize=(12, 12))

plt.subplot(221) 
plt.bar(wells, cells)
plt.subplot(222)
plt.scatter(wells, cells)
plt.subplot(223)
plt.plot(wells, cells)
plt.suptitle('Multiple Plots')
plt.show()
```

Another method to define Multiple Plots side by side

```python
#Initialize the plots

fig = plt.figure(figsize=(16,6))
ax1 = fig.add_subplot(131)
ax1.set(title='vertical bar', xlabel='Well #', ylabel='# cells')

ax2 = fig.add_subplot(132)
ax1.set(title='horizontal bar', xlabel='Well #', ylabel='# cells')

ax3 = fig.add_subplot(133)

ax1.bar(wells, cells)
ax2.barh(wells, cells)
ax3.plot(wells, cells)

plt.show()
```
