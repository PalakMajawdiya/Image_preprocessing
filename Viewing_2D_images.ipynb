**Displaying image using skimage **

```python
from skimage import io

img = io.imread(r'/content/drive/MyDrive/IMRE9.png')

io.imshow(img)
```

**Displaying image using matplotlib**

```python
import matplotlib.pyplot as plt

plt.imshow(img)
```

cmap is color map which is basically used to convey a message so if there is some meaning in your image then you can use this function

```python
plt.imshow(img , cmap = 'hot')

#here image is in RGB so not going to do anything
```

as_grey is used to make your array of pixels in folat values otherwise it is of unsigned int

```python
img_grey = io.imread(r'/content/drive/MyDrive/IMRE9.png' , as_gray = True)

plt.imshow(img_grey , cmap = "hot")

```

```python
plt.imshow(img_grey , cmap = "jet")
```

**Multiple plots using pyplot**

```python
fig = plt.figure(figsize=(10,10))

ax1 = fig.add_subplot(2,2,1) # here 2X2 is a grid and locate it on 1st position 
ax1.imshow(img_grey, cmap = "hot")
ax1.title.set_text('1st')

ax2 = fig.add_subplot(2,2,2) #2X2 at position 2nd
ax2.imshow(img_grey , cmap = 'jet')
ax2.title.set_text('2nd')


ax3 = fig.add_subplot(2,2,3)
ax3.imshow(img_grey , cmap = 'gray')
ax3.title.set_text('3rd')

ax4 = fig.add_subplot(2,2,4)
ax4.imshow(img_grey , cmap = 'nipy_spectral')
ax4.title.set_text('4th')

plt.show()
```

**Using OpenCV library **

cv2.waitkey() is a function for how long you want to display the window 
cv2.waitkey(0) means untill i close the window make it open
cv2.waitkey(3000) it opens the window for 3000 ms that is for 3 sec

```python
import cv2

gray_img = cv2.imread(r'/content/drive/MyDrive/IMRE9.png', 0)
color_img = cv2.imread(r'/content/drive/MyDrive/IMRE9.png', 1)

cv2.imshow("pic from skimage import", img)  
cv2.imshow("color pic from opencv", color_img)
cv2.imshow("gray pic from opencv", gray_img)

cv2.waitKey(0) 
cv2.destroyAllWindows() 
```
